```python
import torch
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.cm as cm
from utils_training import GNN
from utils_training import getLoader
import numpy as np
import matplotlib

from custom_colormaps import create_colormap
colors = [(167,212,121),
            (236,197,244),
            (255,146,143),
            (162,224,223)]
my_cmap = create_colormap(colors, bit=True)
font = {# 'family' : 'Times New Roman',
        # 'weight' : 'bold',
        'size'   : 15}
matplotlib.rc('font', **font)

for i, predict in enumerate([45, 90, 160, 200, 250]):
    model = GNN(n=16,k=4,top=12)
    weights = torch.load(f'C:\\Users\\Gyu-Jang Sim\\Documents\\FIPGraph\\trained_weights\\30_frac100weights_n_16_k_4_avg.pt')
    loader_dir = f'C:\\Users\\Gyu-Jang Sim\\Documents\\FIPGraph\\datalist\\{predict}single.pickle'


    model.load_state_dict(weights)

    model.eval()

    loader = getLoader(datalist_dir=loader_dir, batch_fraction=1)
    iterer = iter(loader)
    batch = next(iterer)

    y = model(batch)

    XX = batch.fip.detach().numpy()
    YY = y.detach().numpy()

    # 0.00265817,0.0009603592368511412
    # XX = XX*0.0009603592368511412 + 0.00265817
    # YY = YY*0.0009603592368511412 + 0.00265817

    plt.figure(1, figsize=(10,10))
    step=0.05
    indices = np.arange(0,1+step,step)
    CDF = pd.DataFrame({'dummy':XX[:,0]})['dummy'].quantile(indices)
    plt.plot(CDF,indices,linestyle=':',color=cm.hsv(i/5),label=f'sve{predict} - CP')

    plt.figure(1)
    CDF = pd.DataFrame({'dummy':YY[:,0]})['dummy'].quantile(indices)
    plt.plot(CDF,indices,color=cm.hsv(i/5),label=f'sve{predict} - GNN')

    plt.legend()


    ###############

    plt.figure(2, figsize=(10,10))
    step=0.05
    indices = np.arange(0,1+step,step)
    CDF = pd.DataFrame({'dummy':XX[:,0]})['dummy'].quantile(indices)
    plt.plot(CDF,indices,linestyle=':',color=cm.hsv(i/5),label=f'sve{predict} - CP')

    plt.figure(2)
    CDF = pd.DataFrame({'dummy':YY[:,0]})['dummy'].quantile(indices)
    plt.plot(CDF,indices,color=cm.hsv(i/5),label=f'sve{predict} - GNN')

    plt.xlim(1,4.5)
    plt.ylim(0.95,1)

    plt.legend()


    ###############


#     bw = 0.00265817 * 20

    plt.figure(3, figsize=(10,10))
    s = pd.Series(YY[:,0])
    # s.plot.kde(bw_method=bw,color=cm.hsv(i/5),label=f'sve{predict}GNN')
    s.plot.kde(color=cm.hsv(i/5),label=f'sve{predict}GNN')

    plt.figure(3, figsize=(10,10))
    s = pd.Series(XX[:,0])
    # s.plot.kde(bw_method=0.00265817,linestyle=':',color=cm.hsv(i/5),label=f'sve{predict}CP')
    s.plot.kde(linestyle='dotted',color=cm.hsv((i-0.1)/5),label=f'sve{predict}CP')

    plt.xlim(-4,4)
    plt.ylim(0,0.5)

    plt.legend()

    ###############


    plt.figure(4, figsize=(10,10))
    s = pd.Series(YY[:,0])
    # s.plot.kde(bw_method=bw,color=cm.hsv(i/5),label=f'sve{predict}GNN')
    s.plot.kde(color=cm.hsv(i/5),label=f'sve{predict}GNN')

    plt.figure(4, figsize=(10,10))
    s = pd.Series(XX[:,0])
    # s.plot.kde(bw_method=bw,linestyle=':',color=cm.hsv(i/5),label=f'sve{predict}CP')
    s.plot.kde(linestyle='dotted',color=cm.hsv((i-0.1)/5),label=f'sve{predict}CP')

    plt.xlim(1.5,4)
    plt.ylim(0,0.1)

    plt.legend()
```

```python
from utils_training import GNN

def get_n_params(model):
    pp=0
    for p in list(model.parameters()):
        nn=1
        for s in list(p.size()):
            nn = nn*s
        pp += nn
    return pp

model = GNN(n=16,k=5,top=3)
print(get_n_params(model))

model = GNN(n=32,k=4,top=4)
print(get_n_params(model))

model = GNN(n=16,k=4,top=12)
print(get_n_params(model))
```
